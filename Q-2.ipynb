```python
import numpy as np
import pandas as pd

import xlwt


import matplotlib.pyplot as plt
%matplotlib nbagg

data = pd.read_csv('facebookdata.csv')
#data = pd.read_csv('C:\Users\USER\Desktop\sem 5\Data\Assignment2')

array= data.values
friend= array[:,0]
post= array[:,1]
like= array[:,2]

#print(sum(friend)/len(friend), round(np.std( friend), 4))
#print(sum(like)/len(like) , round(np.std( like), 4))
#print(sum(post)/len(post) ,round(np.std( post) , 4))



plt.subplot(311)                                                   #histogram of 1st and 15th day 
plt.hist2d(friend,post, bins=100 )
plt.ylabel('Posts')
plt.xlabel('Friends')

plt.title('Number of Friends vs Number of Posts') 
 

plt.subplot(312)                                                   #histogram of every day temprature 
plt.hist2d(friend,like, bins=100 )

plt.ylabel('Likes')
plt.xlabel('Friends')
plt.title('Number of Friends vs Number of Likes')


plt.subplot(313)                                                   #histogram of every day temprature 
plt.hist2d(post,like, bins=100 )

plt.ylabel('Likes')
plt.xlabel('Posts')
plt.title('Number of posts vs Number of Likes')


plt.show()
```
