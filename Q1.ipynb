```python
import numpy as np                        #importing libraries 
import pandas as pd

import xlwt


import matplotlib.pyplot as plt
%matplotlib nbagg

data = pd.read_csv('facebookdata.csv')   #importing Data from .csv file

array= data.values                       #comvering data int numpy array

friend= array[:,0]                       #creating seprate list for each parameter
post= array[:,1]
like= array[:,2]
```

```python
#finding the mean and standard deviation of different parameters

print('Mean and Standard Deviation for Number of Friends are', sum(friend)/len(friend) ,'and', round(np.std( friend),4),'respectively')
print('Mean and Standard Deviation for Number of Posts are', sum(post)/len(post) ,'and', round(np.std(post),4),'respectively')
print('Mean and Standard Deviation for Number of Likes are', sum(like)/len(like) ,'and', round(np.std(like),4),'respectively')
```

```python
plt.figure(1)               #Plotting the graph of 1d histogram                    

plt.subplot(311)                                                   
plt.hist(friend,bins=30)    #dividing the data into 30 equal bins
plt.xlabel('Friends')
plt.ylabel('Frequency')
plt.title('Histogram: Number of Friends') 
 

plt.subplot(312)                                                 
plt.hist(post,bins=30)
plt.xlabel('Posts')
plt.ylabel('Frequency')
plt.title('Histogram: Number of posts')
                                    

plt.subplot(313)   
plt.hist(like,bins=30)
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.title('Histogram: Numbar of Likes')
plt.show()
```
