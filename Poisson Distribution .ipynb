```python
import matplotlib.pyplot as plt
% matplotlib inline
plt.title('Poission Distribution')
plt.xlabel('n')
plt.ylabel('poission value')
import math
i=range(1,41)
p1=4
p2=8
from math import factorial
from math import exp
y1=[]
y2=[]
for elem in i:
    y1.append((p1**elem)*(exp(-p1))/factorial(elem))
    y2.append((p2**elem)*(exp(-p2))/factorial(elem))
plt.scatter(i,y1,label=('lamda=4'),color='r')
plt.scatter(i,y2,label=('lamda=8'),color='g')
plt.legend()
plt.show()
```
