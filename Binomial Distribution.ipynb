```python
import matplotlib.pyplot as plt
% matplotlib inline
import math
plt.title('Binomial Distribution')
plt.xlabel('n')
plt.ylabel('Binomial function value')
i=range(1,31)
j=range(1,61)
k=range(1,91)
p1=0.4
n1=30
p2=0.4
n2=60
p3=0.4
n3=90
from math import factorial
y1=[]
y2=[]
y3=[]
for elem in i:
    y1.append((factorial(n1)/(factorial(n1-elem)*factorial(elem)))*(p1**elem)*((1-p1)**(n1-elem)))
    #plt.plot(i,y1, label='Plot1')
for ele in j:    
    y2.append((factorial(n2)/(factorial(n2-ele)*factorial(ele)))*(p2**ele)*((1-p2)**(n2-ele)))
for el in k:    
    y3.append((factorial(n3)/(factorial(n3-el)*factorial(el)))*(p3**el)*((1-p3)**(n3-el)))
plt.scatter(i,y1, label='Plot1',color='red')
plt.scatter(j,y2, label='plot2')
plt.scatter(k,y3, label='plot3',color='green')
plt.legend()
plt.show()
#print(len(x))
    

    
```
